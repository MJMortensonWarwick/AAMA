# Building a Graph in NetworkX
This tutorial will help us build a graph in NetworkX. The package is in-built in Colab ... on your own machine you may need to install first of all. Here we can just import and start building!

```python
import networkx as nx
import matplotlib.pyplot as plt

g = nx.Graph()
```

Here we have built an empty graph called _g_. Now we can start adding our nodes:

```python
g.add_node(1)

# print to screen
plt.figure(figsize=(10, 10))
nx.draw_networkx(g, with_labels = True)
```

Beautiful! Well a bit boring actually with our single _isolated_ node. Let's give him some friends:

```python
# add nodes from an iterable/list
g.add_nodes_from([2, 3])

# print to screen
plt.figure(figsize=(10, 10))
nx.draw_networkx(g, with_labels = True)
```

We can also add nodes with metadata attached to them like so:

```python
g.add_nodes_from([
    (4, {"name": "Vinh"}),
    (5, {"name": "Wenjuan"}),
])

# print to screen
plt.figure(figsize=(10, 10))
nx.draw_networkx(g, with_labels = True)
```

Our metadata has been saved to the graph, but is not part of the visual display. However, our graph is coming along but could really benefit from some edges about now!

```python
# adding a single edge
g.add_edge(1, 2)

# adding edges from an iterable/list
g.add_edges_from([(2, 4), (1, 3)])

# print to screen
plt.figure(figsize=(10, 10))
nx.draw_networkx(g, with_labels = True)
```

Poor Wenjuan out on her own. However, Vinh has made some friends. Although we can see the graph, we could also generate the node and edge counts as a statistic:

```python
print(f"Number of nodes is: {g.number_of_nodes()}")
print(f"Number of edges is: {g.number_of_edges()}")
```

And the numbers check out! But what if we want to remove a node? Let's remove Vinh:

```python
g.remove_node(4)

# print to screen
plt.figure(figsize=(10, 10))
nx.draw_networkx(g, with_labels = True)
```

And if we want to kill all the nodes:

```python
g.clear()

# print to screen
plt.figure(figsize=(10, 10))
nx.draw_networkx(g, with_labels = True)
```

Let's change _g_ to a directed graph:

```python
# directed graph
dg = nx.DiGraph() # directed graph

# add a directed edge from 1 to 2 and 3 to 1
# give each of them weights
dg.add_weighted_edges_from([(1, 2, 0.5), (3, 1, 0.75)])

# print to screen
plt.figure(figsize=(10, 10))
nx.draw_networkx(dg, with_labels = True)
```

## Loading a graph from Pandas
We can also use Pandas to prepare our data before importing:

```python
import pandas as pd

# build the data as an adjacency matrix
df = pd.DataFrame([[0, 1, 1], [1, 0, 0], [1, 0, 0]])

df.head()
```

As per the matrix, we have zeros on the diagonal (no nodes connected to themselves) and undirected edges joining 0 and 1 plus 0 and 2. We can now build the graph from this:

```python
dfg = nx.from_pandas_adjacency(df)

# print to screen
plt.figure(figsize=(10, 10))
nx.draw_networkx(dfg, with_labels = True)
```
